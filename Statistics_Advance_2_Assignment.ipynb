# **Theory Questions**

1. **What is hypothesis testing in statistics?**
- Hypothesis testing is a statistical method used to make inferences about a population based on sample data. It helps determine whether there is enough evidence to support a particular claim.

2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**
- Null Hypothesis:
  - The default assumption that there is no effect, no difference, or no relationship in the population.
  - It serves as a baseline for statistical testing.
  - Example: The average test score of students is 75.
- Alternative Hypothesis:
  - Represents what the researcher aims to prove—suggesting that a significant difference or effect exists.
  - Example: The average test score is not 75.

3. **What is the significance level in hypothesis testing, and why is it important?**
- The significance level in hypothesis testing represents the probability of rejecting the null hypothesis when it is actually true. It sets the threshold for determining whether an observed effect is statistically significant.
- Importance:
  - Controls error rate: A lower reduces the risk of false positives.
  - Determines statistical significance: If the p-value is less than level of significance, we reject the null hypothesis.
  - Balances confidence vs. risk: A smaller level of significance provides stronger evidence but makes it harder to detect real effects.

4. **What does a P-value represent in hypothesis testing?**
- The P-value in hypothesis testing represents the probability of obtaining results at least as extreme as the observed ones, assuming the null hypothesis is true. It helps determine whether there is enough statistical evidence to reject the null hypothesis.

5. **How do you interpret the P-value in hypothesis testing?**
- The P-value in hypothesis testing indicates the probability of obtaining results at least as extreme as the observed ones, assuming the null hypothesis is true. It helps determine whether there is enough statistical evidence to reject the null hypothesis.
- How to Interpret P-Value:
  - If P-value < significance level, reject null hypothesis, means the observed result is unlikely under null hypothesis, supporting the alternative hypothesis.
  - If P-value > significance level, fail to reject null hypothesis, means the observed data does not provide strong enough evidence to reject the null hypothesis.

6. **What are Type 1and Type 2 errors in hypothesis testing?**
- Type 1:
  - Occurs when we incorrectly reject null hypothesis when it is actually true.
  - It means detecting an effect or difference that does not exist.
  - Example: A medical test incorrectly indicates that a healthy person has a disease.
- Type 2:
  - Occurs when we fail to reject null hypothesis when it is actually false.
  - It means missing a real effect or difference.
  - Example: A medical test fails to detect a disease in a sick person.

7. **What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
- One-Tailed Test:
  - Tests for an effect in only one direction (greater or less than).
  - Used when prior knowledge or theory suggests a specific direction.
  - Example:
    - Null Hypothesis: The average height of students is 170 cm.
    - Alternative Hypothesis: The average height is greater than 170 cm.
- Two-Tailed Test:
  - Tests for an effect in both directions (different but unspecified)
  - Used when there is no prior assumption about direction.
  - Example:
    - Null Hypothesis: The average height of students is 170 cm.
    - Alternative Hypothesis: The average height is different from 170 cm.

8. **What is the Z-test, and when is it used in hypothesis testing?**
- The Z-test is a statistical test used in hypothesis testing when comparing a sample mean to a known population mean. It determines whether the difference between the sample and population means is statistically significant, assuming the sample follows a normal distribution.
- Z-test is used when:
  - The Z-test is most appropriate when the sample size is large because of the Central Limit Theorem, which ensures normality.
  - The test requires knowledge of the population standard deviation to calculate the Z-score.
  - It is used to test whether a sample mean is significantly different from a population mean.

9. **How do you calculate the Z-score, and what does it represent in hypothesis testing?**
- The Z-score (or standard score) measures how far a data point is from the mean, expressed in terms of standard deviations. It is crucial in hypothesis testing because it helps assess whether the observed data significantly differs from the expected value.
- Formula for Z-score
$$ Z = \frac{X - \mu}{\sigma} $$ Where:
  - X: Data point (or sample mean)
  - μ: Population mean
  - σ: Population standard deviation

- Interpretation in Hypothesis Testing
  - Positive Z-score: Data is above the mean.
  - Negative Z-score: Data is below the mean.
  - Z = 0: Data matches the mean exactly.



10. **What is the T-distribution, and when should it be used instead of the normal distribution?**
- The T-distribution, also called Student’s t-distribution, is a probability distribution that is used in hypothesis testing and confidence intervals when the sample size is small and/or the population standard deviation is unknown.
- When to Use the T-Distribution Instead of the Normal Distribution:
  - Small Sample Size (( n < 30 ).
  - When ( n ) is small, the T-distribution accounts for the added uncertainty.
  - Unknown Population Standard Deviation (( \sigma )).
  - If ( \sigma ) is unknown and must be estimated from sample data, the T-distribution is used.

11. **What is the difference between a Z-test and a T-test?**
- Z-test:
  - Used for large samples (( n >= 30 )).
  - Requires known population standard deviation (( \sigma )).
  - Based on the normal distribution.

- T-test:
  - Used for small samples (( n < 30 )).
  - Used when population standard deviation is unknown (uses sample standard deviation ( s )).
  - Based on the T-distribution, which has wider tails to account for uncertainty.





12. **What is the T-test, and how is it used in hypothesis testing?**
- The T-test is a statistical test used in hypothesis testing to determine whether there is a significant difference between the means of two groups. It is particularly useful when sample sizes are small (( n < 30 )) and the population standard deviation is unknown.
- T-test can be used in following steps:
  - Step1: Find the null hypothesis and alternate hypothesis.
  - Step2: find degree of freedom(dof) = n-1 and level of significance (σ).
  - Step3: Calculate t-statistics = (x-μ)/s/√n
  - Step4: Calculate t-critical corresponding to σ.
  - Step5: if t-stats < t-critical
    - failed to reject null hypothesis.
  - else:
    - reject the null hypothesis.

13. **What is the relationship between Z-test and T-test in hypothesis testing?**
- The Z-test and T-test are closely related statistical methods used in hypothesis testing to determine whether sample data provides enough evidence to reject the null hypothesis.
- How They Relate
  - Both tests compare means to determine statistical significance.
  - The T-test generalizes the Z-test by accounting for the uncertainty in estimating the population standard deviation.
  - As sample size increases, the T-distribution approaches the normal distribution, making the Z-test appropriate.
  - If ( n < 30 ) or ( \sigma ) is unknown, use the T-test instead of the Z-test.

14. **What is a confidence interval, and how is it used to interpret statistical results?**
- A confidence interval (CI) is a range of values that is likely to contain the true population parameter (such as a mean or proportion) with a certain level of confidence. It provides a way to quantify the uncertainty in statistical estimates.
- Confidence Intervals Are Used in Interpretation by:
  - A narrow confidence interval indicates a more precise estimate.
  - A wide confidence interval suggests greater uncertainty in the measurement.
  - If a confidence interval excludes the null hypothesis value, it suggests statistical significance, leading to rejection of null hypothesis.
  - If the interval includes the null hypothesis value, we fail to reject the null hypothesis, meaning no strong evidence against it.

15. **What is the margin of error, and how does it affect the confidence interval?**
- The margin of error (MoE) is the range above and below a sample estimate that expresses the uncertainty in a statistic. It helps determine how precise the estimate is in relation to the true population parameter.
- It affect  the confidence  interval by:
  - A larger margin of error leads to a wider confidence interval, meaning less precise estimates.
  - A smaller margin of error leads to a narrower confidence interval, meaning more precise estimates.

19. **What is the F-distribution, and when is it used in hypothesis testing?**

20. **What is an ANOVA test, and what are its assumptions?**
- ANOVA is a statistical test used to determine whether there are significant differences between the means of three or more groups. Instead of performing multiple t-tests, ANOVA compares variances within and between groups to check for differences.
- ANOVA requires these assumptions:
  - Observations are independent of each other (no repeated measures in One-Way ANOVA).
  - Data should be normally distributed within each group.
  - Each group should have equal variance.

21. **What are the different types of ANOVA tests?**
- One-Way ANOVA:
  - Compares means across one independent variable (e.g., test scores in three schools).
- Two-Way ANOVA:
  - Examines effects of two independent variables (e.g., test scores based on teaching method and gender).
- Repeated Measures ANOVA:
  - Compares means from the same group over different time points (e.g., blood pressure before, during, and after treatment).

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
import pandas as pd
import scipy.stats as st

sample_data = [50, 52, 48, 53, 51, 50, 54, 47, 61, 49]
population_mean = 50
population_std = 20
alpha = 0.05
n = len(sample_data)

sample_mean = np.mean(sample_data)

#calculate z-score
z_score = (sample_mean - population_mean)/(population_std/np.sqrt(n))

#calculate p-value
#Two Tail Test

p_value = 2*(1-st.norm.cdf(abs(z_score)))

if p_value < alpha:
  print("Reject the null Hypothesis")
else:
  print("Failed to reject the null hypothesis")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


pop_mean = 100
pop_std = 15
sample_size = 30

sample_data = np.random.normal(loc=105, scale=pop_std, size=sample_size)


sample_mean = np.mean(sample_data)


z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))


p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))


alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)

if abs(z_score) > critical_value:
    print("Reject the null hypothesis (H₀): The sample mean significantly differs from the population mean.")
else:
    print("Fail to reject the null hypothesis (H₀): No significant difference.")


x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(8,6))
plt.plot(x, y, label="Standard Normal Distribution", color="blue")
plt.axvline(-critical_value, color="red", linestyle="dashed", label="Critical Value (-Z)")
plt.axvline(critical_value, color="red", linestyle="dashed", label="Critical Value (+Z)")
plt.axvline(z_score, color="green", linestyle="dashed", label=f"Observed Z-score ({z_score:.2f})")
plt.fill_between(x, y, where=(x < -critical_value) | (x > critical_value), color='red', alpha=0.3)
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.title("Decision Region in Two-Tailed Z-Test")
plt.legend()
plt.grid()
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(alpha=0.05, beta=0.20, mu_0=100, mu_1=105, sigma=15, n=30):

    se = sigma / np.sqrt(n)
    critical_value = mu_0 + stats.norm.ppf(1 - alpha) * se

    x = np.linspace(mu_0 - 4*sigma, mu_1 + 4*sigma, 1000)

    y_null = stats.norm.pdf(x, mu_0, se)
    y_alt = stats.norm.pdf(x, mu_1, se)


    beta_area = stats.norm.cdf(critical_value, mu_1, se)


    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label="Null Distribution (H₀)", color="blue")
    plt.plot(x, y_alt, label="Alternative Distribution (H₁)", color="red")


    plt.fill_between(x, y_null, where=(x > critical_value), color='blue', alpha=0.3, label="Type 1 Error (Alpha)")


    plt.fill_between(x, y_alt, where=(x < critical_value), color='red', alpha=0.3, label="Type 2 Error (Beta)")

    plt.axvline(critical_value, color="black", linestyle="dashed", label=f"Critical Value ({critical_value:.2f})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.title("Visualization of Type 1 and Type 2 Errors")
    plt.legend()
    plt.grid()
    plt.show()

    print(f"Critical Value: {critical_value:.2f}")
    print(f"Type 2 Error Probability (Beta): {beta_area:.4f}")


visualize_errors()
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
def confidance_interval(sample_data, confidance = 0.95, population_std = None):
  size = len(sample_data)
  mean = np.mean(sample_data)
  std = np.std(sample_data)

  if population_std:
    z_score = stats.norm.ppf(1-(1-confidance)/2)
    margin_of_error = z_score * (population_std/np.sqrt(size))

  else:
    t_score = stats.t.ppf(1-(1-confidance)/2, df = size-1)
    margin_of_error = t_score * (std/np.sqrt(size))

  return mean-margin_of_error, mean+margin_of_error

sample_data = np.random.normal(loc = 20, scale = 25, size = 50)
low, up = confidance_interval(sample_data)
print(f"The confidance interval of the sample data is ({low:.2f},{up:.2f})")
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
import scipy.stats as stats
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)


var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)


F_statistic = var1 / var2


df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H₀): Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis (H₀): No significant difference in variances.")
```

22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

sample1 = np.random.normal(loc=50, scale=5, size=30)
sample2 = np.random.normal(loc=55, scale=5, size=30)
sample3 = np.random.normal(loc=60, scale=5, size=30)

f_statistic, p_value = stats.f_oneway(sample1, sample2, sample3)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H₀): At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis (H₀): No significant difference between group means.")

plt.figure(figsize = (10,6))
sns.boxplot(data = [sample1, sample2, sample3])
plt.title("Group Means")
plt.xlabel("Samples/Groups")
plt.ylabel("Values")
plt.show()
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
n = 100
sample_std = 20
sample_mean = 70
sample = np.random.normal(loc = sample_mean, scale = sample_std, size = n)

sample_var = np.var(sample, ddof=1)
pop_var = 100

chi2_stat = (n-1)*sample_var/pop_var

p_value = 2*min(stats.chi2.cdf(chi2_stat, df = n-1), 1-stats.chi2.cdf(chi2_stat, df = n-1))

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H₀).")
else:
    print("Fail to reject the null hypothesis (H₀).")
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
import scipy.stats as stats


successes_group1 = 150
total_group1 = 1000

successes_group2 = 180
total_group2 = 1200


p1 = successes_group1 / total_group1
p2 = successes_group2 / total_group2


p_combined = (successes_group1 + successes_group2) / (total_group1 + total_group2)
se = np.sqrt(p_combined * (1 - p_combined) * (1/total_group1 + 1/total_group2))


z_score = (p1 - p2) / se


p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H₀).")
else:
    print("Fail to reject the null hypothesis (H₀).")
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=50, scale=15, size=100)

var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

F_statistic = var1 / var2
df1 = len(data1) - 1
df2 = len(data2) - 1
p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H₀): The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis (H₀): No significant difference in variances.")


plt.figure(figsize=(8,6))
plt.hist(data1, bins=15, alpha=0.6, label="Dataset 1 (Std=10)", color="blue", density=True)
plt.hist(data2, bins=15, alpha=0.6, label="Dataset 2 (Std=15)", color="red", density=True)
plt.axvline(np.mean(data1), color="blue", linestyle="dashed", linewidth=2, label=f"Mean Dataset 1 ({np.mean(data1):.2f})")
plt.axvline(np.mean(data2), color="red", linestyle="dashed", linewidth=2, label=f"Mean Dataset 2 ({np.mean(data2):.2f})")
plt.title("Comparison of Two Normally Distributed Datasets")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
